## Introducción a línea de comandos

Es muy probable que ya tengas experiencia utilizando interfaces gráficas de usuario (GUI). Estas interfaces —cuya pronunciación habitual es "güi"— facilitan la interacción con tu computadora mediante elementos visuales. Piensa en la acción de hacer clic en un icono para abrir tu navegador web preferido: eso es una GUI en acción. Las GUI destacan por su facilidad de uso, pero en el mundo del desarrollo de software, también trabajamos con otro tipo de interfaz que conoceremos a continuación.

A diferencia de las intuitivas interfaces gráficas de usuario (GUI), existe otro método especialmente relevante en el mundo del desarrollo de software: la interfaz de línea de comandos (CLI). La CLI, se basa en la introducción de comandos de texto para indicar al sistema operativo qué acciones debe realizar.
¿Por qué, entonces, deberías aprender a utilizar la CLI como futuros profesionales de la ciencia o análisis de datos?

En primer lugar, la CLI es fundamental para la gestión eficiente de entornos y dependencias de proyectos. Ya sea creando entornos virtuales aislados para evitar conflictos entre proyectos, o gestionando las versiones de las librerías que utilizas con pip. La línea de comandos te ofrece un control directo y preciso sobre estos aspectos cruciales del desarrollo y la ciencia de datos.

En segundo lugar, la CLI es indispensable para la automatización de tareas a nivel del sistema operativo. Esto va más allá de la automatización dentro de un lenguaje de programación como Python. Imagina tener que mover, copiar o renombrar grandes cantidades de archivos de datos, o ejecutar scripts periódicamente para la recopilación y procesamiento inicial de información. Conocer el funcionamiento de la CLI, te permite escribir scripts que automaticen estas tareas a un nivel fundamental del sistema, independientemente del lenguaje de programación que utilices después para el análisis.

Finalmente, la CLI te proporciona una comprensión más profunda de cómo funciona tu sistema operativo. Al interactuar directamente con él a través de comandos, entiendes mejor la estructura de archivos, los procesos en ejecución y cómo se gestionan los recursos. Este conocimiento subyacente es valioso para solucionar problemas, optimizar el rendimiento de tu código y comprender el entorno en el que se ejecutan tus aplicaciones o análisis de datos.

Habrá un montón de nuevo material para aprender. No te preocupes si no lo entiendes todo desde el principio. La idea principal es proporcionarte conocimientos y práctica con la CLI que te preparará para los escenarios más comunes. Con esto en mente, ¡comencemos nuestro viaje, que durará unas 5 horas!

## Entornos de programación/desarrollo

Imagina que te preparas para embarcarte en un emocionante proyecto, ya sea desentrañar los secretos ocultos en vastos conjuntos de datos o construir una aplicación innovadora que transforme ideas en realidad. Al igual que un artesano necesita un taller bien equipado o un científico un laboratorio meticulosamente organizado, tú, como futuro analista o científico de datos, necesitas un espacio de trabajo digital adecuado. Este espacio, conocido como entorno de programación o entorno de desarrollo, es fundamental para convertir tus ideas en código funcional y resultados impactantes. Entender la naturaleza y las diferencias entre estos términos te permitirá optimizar tu flujo de trabajo y maximizar tu productividad.

### Entorno de programación vs. entorno de desarrollo: dos niveles de preparación

Aunque a menudo se utilicen indistintamente, podemos entenderlos como dos niveles de preparación en tu espacio de trabajo digital.

* El entorno de programación representa los componentes esenciales que te permiten escribir y ejecutar código en un lenguaje específico. Piensa en él como el kit básico de herramientas. Para Python, esto incluye fundamentalmente un editor de texto donde plasmar tus ideas en código y un intérprete de Python que traduce esas instrucciones para que la computadora las comprenda y ejecute. Es el punto de partida, la base sobre la cual construyes.
* El entorno de desarrollo es una extensión más completa de este kit básico. Abarca el entorno de programación, pero añade un conjunto de herramientas y utilidades adicionales diseñadas para facilitar el ciclo de vida completo del análisis de datos a gran escala. Imagina que a tu kit básico de herramientas le añades un banco de trabajo organizado, instrumentos de medición precisos y guías especializadas.

### Tu kit básico: el entorno de programación

En su forma más fundamental, un entorno de programación para Python se centra en dos elementos clave:

* El Editor de Texto es la aplicación donde escribes tus "instrucciones" en lenguaje Python. Si bien cualquier editor de texto básico puede servir, es crucial evitar procesadores de texto como Word o Google Docs, ya que introducen formatos que pueden corromper la sintaxis limpia que Python necesita. Editores de código especializados como VS Code, Sublime Text o Atom ofrecen ventajas significativas, como el resaltado de sintaxis (colores que ayudan a identificar el código), lo que facilita la detección de errores y mejora la legibilidad. Piensa en el editor de texto como tu lienzo digital donde tus ideas toman forma.
    El Intérprete de Python es el traductor de tus Ideas. Una vez que has escrito tu código en un archivo (guardado con la extensión .py para que el sistema y otros programas lo identifiquen como código Python), necesitas un programa que pueda leer, interpretar y ejecutar esas instrucciones. Ese es el intérprete de Python. Es el "cerebro" que entiende lo que has escrito y le dice a la computadora qué hacer.

### El entorno de desarrollo

Un entorno de desarrollo va un paso más allá, proporcionando un conjunto de herramientas que agilizan y mejoran significativamente el proceso de creación y análisis.
La importancia de los entornos de desarrollo

En el vasto universo de la programación, un entorno de desarrollo es un espacio aislado y configurado específicamente para escribir, probar y ejecutar código. Imagina tener un laboratorio científico dedicado a un experimento en particular. Este laboratorio tiene todas las herramientas, los reactivos y las condiciones controladas necesarias para llevar a cabo la investigación sin interferencias externas. Un entorno de desarrollo cumple una función similar en el mundo digital.

Uno de las superpoderes de los entornos de desarrollo es su capacidad para crear aislamiento. Piensa en cada proyecto como un universo independiente, con sus propias reglas y necesidades. Un proyecto podría requerir una versión específica de una librería de Python, mientras que otro necesita una versión completamente diferente. Sin entornos, estas dependencias podrían chocar, causando errores y frustración. Los entornos de desarrollo actúan como fronteras invisibles, asegurando que cada proyecto tenga exactamente las herramientas y versiones que necesita, sin afectar a los demás. Es como tener múltiples laboratorios, cada uno equipado para un experimento único.
Experimentar sin temor a romper nada

El aprendizaje y la innovación a menudo implican probar cosas nuevas: una librería recién lanzada, una versión beta de un lenguaje, un enfoque algorítmico diferente. Sin un entorno de desarrollo aislado, esta experimentación podría ser arriesgada, amenazando la estabilidad de tu entorno de trabajo principal. Los entornos te ofrecen un campo de juego seguro, un espacio donde puedes explorar, cometer errores y aprender sin la preocupación de dañar tu configuración base o afectar proyectos importantes. Es como tener un arenero donde los niños pueden construir castillos y destruirlos sin consecuencias para el resto del jardín.
Código que funciona en cualquier lugar

¿Alguna vez has escuchado la frase "funciona en mi máquina"? Es un lamento común en el mundo del desarrollo. Los entornos de desarrollo son la clave para erradicar este problema. Al definir un entorno específico con las versiones exactas de lenguajes y librerías, te aseguras de que tu código no solo funcione en tu computadora, sino también en la de tus colaboradores o en los servidores donde finalmente se ejecute tu proyecto. Es como tener una receta detallada que garantiza que el mismo plato delicioso se pueda preparar en cualquier cocina, siempre y cuando se utilicen los mismos ingredientes y cantidades.
Trabajando juntos en armonía digital

En el mundo actual, la mayoría de los proyectos de software y ciencia de datos son esfuerzos colaborativos. Los entornos de desarrollo actúan como un lenguaje común, asegurando que todos los miembros del equipo estén utilizando las mismas herramientas y versiones. Esto reduce la fricción, facilita la comunicación y minimiza los problemas de integración. Es como tener un conjunto de herramientas estándar en un equipo de construcción, donde todos saben exactamente qué usar y cómo funciona, permitiendo un trabajo en equipo fluido y eficiente

Para los aspirantes a analistas y científicos de datos, comprender y utilizar entornos de desarrollo no es solo una habilidad técnica más, sino una inversión en su futuro. A medida que avances en tu carrera, te encontrarás con proyectos más complejos, colaboraciones en equipo y la necesidad de desplegar modelos y análisis en diferentes plataformas. Dominar los entornos de desarrollo te proporcionará una base sólida para enfrentar estos desafíos con confianza y eficiencia, asegurando que su trabajo sea reproducible, confiable y listo para el mundo real. Es el primer paso para construir un legado digital duradero.

## Instalación y uso de módulos y paquetes

Ahora que sabes cómo configurar y trabajar en un entorno de programación en tu máquina local, es hora de obtener más información sobre los paquetes de Python y su función en tu entorno.

**Módulos, paquetes y librerías**

Imagina que estás construyendo algo con Legos. Empiezas con las piezas básicas, pero pronto te das cuenta de que alguien ya inventó ruedas, ventanas o incluso un motor pequeño. En lugar de tener que diseñar y construir cada una de estas partes desde cero, ¡puedes simplemente tomarlas y usarlas en tu creación!

En Python, los módulos, paquetes y librerías son como esas "piezas" prefabricadas. Son códigos escritos por otras personas que hacen cosas específicas y que tú puedes usar en tus propios programas sin tener que escribir todo desde el principio.

Entendiendo las "Piezas":

* **Módulo**: Piensa en un módulo como un solo archivo de Python (un archivo .py). Dentro de este archivo, hay "instrucciones" (funciones, clases, variables) que te ayudan a resolver un problema en particular. Es como una sola pieza de Lego con una función específica.
* **Paquete**: Un paquete es como una caja que contiene varias "piezas" de Lego relacionadas (varios módulos). Es una forma de organizar muchos archivos de Python que trabajan juntos.
* **Librería**: Una librería es como una caja grande que contiene muchas cajas más pequeñas (paquetes) y también algunas piezas sueltas (módulos). Es una colección grande de código útil para hacer muchas cosas diferentes.

Nota: A veces, la gente usa estas tres palabras como si significaran lo mismo, y está bien. Lo importante es entender que son código útil escrito por otros que puedes usar en tus proyectos.
Cómo utilizar las "piezas": ¡importando código!

Para usar estas "piezas" en tu código Python, necesitas importarlas. Ya hiciste esto antes:

import os

cpu_num = os.cpu_count()
print(f"Mi computadora tiene {cpu_num} núcleos de CPU")

Aquí, os es un módulo que viene con Python y tiene herramientas para hablar con el sistema operativo de tu computadora (como averiguar cuántos "cerebros" tiene). Para usar una de las "instrucciones" dentro de os (como cpu_count()), escribimos el nombre del módulo, un punto (.), y luego el nombre de la instrucción: os.cpu_count().

También has visto esto:

import numpy as np
from matplotlib import pyplot as plt

Aquí, numpy y matplotlib son paquetes muy populares para trabajar con números y gráficos en Python. La parte as np y as plt es como darles un apodo corto. En lugar de escribir numpy.pi para obtener el valor de pi, podemos escribir np.pi. Esto hace que tu código sea más fácil de leer. La mayoría de los paquetes tienen apodos estándar que la gente usa, ¡así que es bueno aprenderlos!

### Asegúrate de que tus "piezas" encajen: chequeo de versiones

Imagina que estás construyendo un modelo complejo con tus Legos. Si intentas usar una rueda de un set antiguo que no encaja bien con las piezas nuevas, ¡tu construcción podría tambalearse o incluso desmoronarse!

En el mundo de Python, los paquetes (nuestras "piezas" prefabricadas) también reciben actualizaciones. A veces, el código que escribiste usando una versión antigua de un paquete podría comportarse de manera extraña o incluso dejar de funcionar con una versión más nueva. Esto puede causar errores inesperados. Por eso, es importante saber qué versión de tus paquetes estás usando.
¿Qué versión tengo?

Para los paquetes que vienen con Python (los "Legos básicos"), su versión generalmente coincide con la versión de Python que tienes instalada. Puedes ver tu versión de Python abriendo la terminal y escribiendo python (esto te mostrará una información que incluye la versión). En Jupyter Notebook, puedes ejecutar este código:

```
    import sys
    sys.version
    
    Esto te dirá qué versión de Python se está utilizando.
    
    '3.8.10 (default, May 19 2021, 13:12:57) [MSC v.1916 64 bit (AMD64)]'
```

Para la mayoría de los otros paquetes que instalas (las "piezas" especiales), puedes averiguar su versión después de importarlos usando algo llamado el atributo __version__. ¡Ojo con los dos guiones bajos antes y después de "version"! Por ejemplo, para ver la versión del paquete pandas (que usamos mucho para trabajar con datos), harías esto:

```
    import pandas as pd
    print(pd.__version__)
    
    1.2.5
```

**Un pequeño truco con las "cajas" de "piezas"**

A veces, importamos solo una parte de una "caja" grande de "piezas". Por ejemplo, de la librería matplotlib (que nos ayuda a hacer gráficos), a menudo solo usamos la parte llamada pyplot:

```
    from matplotlib import pyplot as plt
    print(plt.__version__)
    
    Esto te dará un error! ¿Por qué? Porque la información de la versión pertenece a la "caja" grande (matplotlib), no a la parte pequeña que importamos (pyplot).
    
    ---------------------------------------------------------------------------
    AttributeError                            Traceback (most recent call last)
    <ipython-input-15-8fe02450c2db> in <module>
          1 from matplotlib import pyplot as plt
    ----> 2 print(plt.__version__)
    
    AttributeError: module 'matplotlib.pyplot' has no attribute '__version__'
    
    Para ver la versión de pyplot, ¡necesitamos importar la "caja" completa de matplotlib!:
    
    import matplotlib
    print(matplotlib.__version__)
```

Esto nos mostrará la versión de toda la librería matplotlib, que es la que controla cómo funciona pyplot.

# Capítulo 4 Git y GitHub

## Realizar cambios a un repositorio

Imagina que estás desarrollando un proyecto de análisis de datos con un equipo. Podría ser cualquier cosa: desde la limpieza y visualización de un gran conjunto de datos hasta la creación de un modelo predictivo. Ahora bien, cada miembro del equipo está trabajando en una parte diferente del proyecto.

Suponiendo que la empresa en la que trabajan tiene una cuenta de Github, en donde todos sus desarrollos son almacenados, tu intención es almacenar tu trabajo dentro de el repositorio destinado para ese proyecto en particular.

Para ello, tu flujo de trabajo será el siguiente:

1. Creación de un Fork

Puesto que buscas trabajar en un repositorio que no es tuyo, el primer paso es crear un “fork” o rama de trabajo. Piensa en un "fork" como pedir una copia personal del proyecto para ti, hasta ese punto en el tiempo. Esta copia se guarda en tu propia cuenta de GitHub, permitiéndote experimentar y hacer cambios sin afectar el repositorio original.
     
     
2. Clonar Repositorio.

Una vez que tienes tu propia copia (o "fork") del repositorio en GitHub, necesitas descargarlo a tu computadora para poder trabajar en él. Para esto, usas el comando git clone

```git clone [URL_DEL_REPOSITORIO]```

Ahora bien, si ya cuentas con una copia del repositorio en tu computadora, pero quieres asegurarte de que esté al día con los últimos cambios realizados por tu equipo,  puedes hacerlo de la siguiente forma. 

```git pull```

El comando git pull "tira" o trae las últimas actualizaciones del repositorio remoto a tu máquina local. Puedes pensar en git clone como la primera vez que obtienes el repositorio, y git pull como la forma de mantenerlo actualizado después de eso.

* Trabajo en tu proyecto :Ahora que tienes el código en tu máquina, puedes editar los archivos como lo harías normalmente. Abre tus editores de texto o entornos de desarrollo, haz todos los cambios necesarios en el código o los datos, y guarda tus modificaciones.

* Revisar el Estado de tus Cambios: Antes de preparar tus cambios, es útil saber qué archivos Git ha detectado que se modificaron o se crearon. Para eso, usamos el comando git status:

```git status```

Este comando te mostrará los archivos que Git reconoce que han sufrido modificaciones o que son nuevos. Los verás en color rojo, indicando que son cambios que aún no han sido preparados para un commit.

* Preparación de los cambios: Una vez que has editado tus archivos, necesitas indicarle a Git qué cambios quieres incluir en la próxima actualización. Esto se hace con git add:

```git add [archivo]```

Por ejemplo, si modificaste un archivo llamado chequeo.txt y quieres incluirlo, escribirías:

```git add chequeo.txt```

Ahora bien, si has realizado muchos cambios y quieres incluir todos los archivos modificados o nuevos en la carpeta actual, puedes usar un atajo:

git add .

El punto (.) le dice a Git que incluya todos los archivos que han sido modificados o que son nuevos en el directorio actual. Después de ejecutar git add (ya sea para un archivo específico o usando git add .), si vuelves a ejecutar git status, verás que estos archivos han cambiado a color verde, indicando que han sido preparados para el siguiente commit.

* Guardar los cambios en el historial: Una vez que has preparado los archivos con git add, el siguiente paso es "confirmar" o "commit" tus cambios. Un "commit" es como crear un "punto de control" en el historial del proyecto. Esto te permite regresar a esta versión específica si alguna vez lo necesitas. Siempre debes incluir un mensaje descriptivo con tu commit para que tú y tu equipo sepan qué cambios se hicieron:

```git commit -m "Descripción de los cambios"```

Por ejemplo, si agregaste una columna con nombres y apellidos, tu commit podría verse así:

```git commit -m "Columna nombre y apellido agregada"```

* Envía los cambios a GitHub: Finalmente, una vez que tus cambios están "comiteados" localmente, necesitas enviarlos a tu repositorio en GitHub para que estén disponibles para tu equipo o para ti mismo desde cualquier lugar. Para esto, usas el comando git push:

```git push```

Este comando sube tus commits desde tu computadora a tu repositorio en GitHub. ¡Ahora, tanto tu copia local como tu repositorio en GitHub están sincronizados con tus últimos cambios!
