## Cosas que un analista necesita saber sobre Internet: navegadores, HTML y HTTP

Tus colegas han compilado una lista de sitios web de los que les gustaría obtener información sobre productos. Tu primera tarea será descargar una página web de un servidor. Sin embargo, no nos apresuremos a escribir consultas de servidor. En primer lugar, deberíamos echar un vistazo a cómo funciona Internet en realidad.

Los primeros prototipos de lo que ahora llamamos Internet aparecieron a principios de la década de 1960, pero pasarían otros 30 años antes de que las cosas realmente se pusieran en marcha. En aquellos primeros días, no había reglas generales para el uso de Internet. Es decir, no tenía interfaz.

En 1991, Tim Berners-Lee, físico, trabajaba en el laboratorio del CERN junto con su equipo. Querían simplificar la interacción con Internet mediante la creación de reglas que hicieran posible que los PC dentro de una red de contacto se comunicaran. Así es como se les ocurrió la idea de la World Wide Web.

**¿Qué tenía que suceder para que Internet adquiriera la forma que tiene hoy?**

Dado que Internet es una red de contacto de PC que intercambia datos, estos datos deben estar estructurados. Debe haber una manera de representar esta información en Internet.

Sin embargo, no basta con simplemente representar información; a los PC se les debe enseñar cómo mostrarla. Había una necesidad de software que permitiera leer y mostrar información en Internet a los usuarios. Además, era necesario que existiera un conjunto de reglas que rigieran cómo los PC envían y reciben esta información.

Como resultado, Tim Berners-Lee y su equipo inventaron:

* Un lenguaje con el que se pueden crear documentos de Internet: HTML (Hypertext Markup Language, o lenguaje de marcado de hipertexto)
* Una aplicación de software para ver dichos documentos: un navegador web
* Reglas generales para transferir documentos: HTTP (Hypertext Transfer Protocol, o protocolo de transferencia de hipertexto)

## Protocolos de transferencia

Internet es una red a través de la cual los PC intercambian información. Para que esto sea posible, debe haber reglas que rijan cómo cada PC envía datos a los demás.

El intercambio de datos en Internet se basa en el principio de "solicitud-respuesta": un navegador genera una solicitud, luego el servidor la analiza y envía una respuesta. Las reglas para formular solicitudes y respuestas están determinadas por lo que se conoce como protocolo de transferencia, en este caso, HTTP. El "HT" significa "hipertexto", ya que el protocolo se diseñó originalmente para transferir documentos HTML. Hoy, sin embargo, se utiliza para transferir todo tipo de datos a través de Internet.

Aquí va una analogía: cuando necesitas enviar una carta, debes ponerla en un sobre, escribir la dirección del destinatario de acuerdo con ciertas reglas, luego sellarlo y llevarlo a la oficina de correos. Los PC y los servidores también intercambian "cartas". El protocolo HTTP es, en el fondo, las reglas para crearlas y enviarlas.
El protocolo HTTPS

La mayoría de los sitios web actuales utilizan un protocolo mejorado de transferencia de datos denominado HTTPS.

image

Fuente: https://transparencyreport.google.com/https/overview

La "S" significa seguro, ya que HTTPS es una versión más segura de HTTP. Este protocolo garantiza que toda la comunicación entre tu navegador y un sitio web esté encriptada.

Los sistemas bancarios en línea y las redes sociales cambiaron a HTTPS hace mucho tiempo. Los navegadores modernos generalmente envían una advertencia de seguridad cuando abres sitios que usan HTTP. Nadie quiere que la información personal de los usuarios esté disponible para otros. Excepto los piratas informáticos, por supuesto.
El navegador y las solicitudes

Cuando accedes a un sitio web, tu navegador envía una solicitud HTTP al servidor. El servidor, a su vez, formula una respuesta: el código HTML de la página correspondiente.

Formato de solicitud

Una solicitud formada por un navegador puede contener lo siguiente:

    Método HTTP: determina la operación a realizar. Existen varios métodos. Los más populares son GET y POST. El primero solicita datos del servidor, mientras que el segundo los envía.
    Ruta: el segmento de la dirección que sigue al nombre del sitio example.com/hello la ruta es /hello).
    Versión HTTP: la versión del protocolo HTTP utilizado para enviar la solicitud (por ejemplo, HTTP/1.1).
    Encabezados de la solicitud: se utilizan para enviar información adicional al servidor.
    Cuerpo de la solicitud: por ejemplo, el cuerpo de una solicitud POST son los datos que se envían. No todas las solicitudes tienen cuerpo.

## Introducción a HTML

HTML significa lenguaje de marcado de hipertexto. El nombre se explica por sí mismo: cada objeto de la página debe marcarse para que se muestre correctamente. Este marcado implica colocar bloques de información dentro de comandos llamados "etiquetas". Estas etiquetas le dicen a los navegadores cómo mostrar la información que rodean.

Un elemento HTML se compone de etiquetas y del contenido que estas contienen. Una etiqueta HTML consta de un nombre rodeado de corchetes angulares. Por ejemplo, la etiqueta del encabezado principal es `<h1>`.

Las etiquetas en un código son como señales de tráfico. Cuando llegas a una ciudad, por ejemplo, verás una señal que dice "Bienvenido a San Francisco", y mientras conduces, verás una que dice "Ahora usted se marcha de San Francisco". De manera similar, cada elemento HTML comienza y termina con etiquetas de apertura y cierre (la etiqueta de cierre comienza con una barra diagonal). Normalmente se hace referencia al elemento resultante por el nombre de su etiqueta.

<image>

Los analistas son como detectives: rastrean ciertas etiquetas y extraen su contenido. Pero, ¿qué etiquetas buscan?

Esta es la estructura típica de una página HTML:

1. <html> ... </html>

La etiqueta <html> presenta cada documento HTML e indica su comienzo, mientras que </html> marca su final. El <head> y el <body> del documento HTML se encuentran entre estas etiquetas.

2. <head> ... </head>

Estas etiquetas marcan el encabezado del documento. Las etiquetas que introducen el título del documento (<title>) y la información meta (adicional) (<meta>) se colocan entre estas etiquetas.

3. <body> ... </body>

La etiqueta <body> marca el comienzo del cuerpo de una página HTML. Todo el contenido de la página (títulos, párrafos de texto, tablas, imágenes) se coloca dentro del cuerpo.

Para que el marcado sea más claro, los desarrolladores dejan comentarios dentro de las etiquetas especiales `<!-- -->` en el código de la página. Esto es muy útil para los analistas, así que dejaremos comentarios en código en nuestros ejemplos.

Los analistas a menudo analizan las tablas. Por lo general, se colocan en los elementos table, entre las etiquetas `<table>` y `</table>`. La etiqueta de apertura `<table>` marca el comienzo de una tabla, y la etiqueta de cierre `</table>` marca su final. Dentro de este elemento, el contenido de la tabla se divide en filas mediante etiquetas `<tr>` (fila de tabla), y las filas, a su vez, se dividen en celdas mediante etiquetas `<td>` (datos de tabla). La primera fila generalmente contiene encabezados de columna en lugar de celdas normales. Se colocan entre las etiquetas `<th>` (encabezado de tabla).

Ejemplo:
Apellido	Nombre	Grupo

Para obtener datos del servidor, usaremos el método `get()`, y para enviar solicitudes HTTP, necesitamos la librería Requests. Importamos la librería:

`import requests`

... ¡y escribimos nuestra primera solicitud GET!

Aquí está la dirección de una página sobre los naufragios más famosos del siglo XX. Guardaremos este enlace en la variable URL. Una URL, como probablemente sepas, es una dirección web. Significa localizador uniforme de recursos.

`URL = 'https://practicum-content.s3.us-west-1.amazonaws.com/new-markets/Data_sprint_6/web-sites/most_famous_shipwrecks_es_1.html?etag=87662559becdeb134637e5670502d6d6'`

El método `get()` de la librería Requests actúa como un navegador. Le pasaremos el enlace como argumento. El método enviará al servidor una solicitud `GET`, luego procesará la respuesta que reciba y devolverá una respuesta, un objeto que contiene la respuesta del servidor a la solicitud.

`req = requests.get(URL) # guardando el objeto de respuesta como variable requerida`

Un objeto response contiene la respuesta del servidor: el código de estado, el contenido de la solicitud y el código de la propia página HTML. Los atributos de los objetos response posibilitan obtener solo los datos relevantes del servidor. Por ejemplo, un objeto response con el atributo text solo devolverá el contenido de texto de la solicitud:

`print(req.text) # el nombre del atributo se coloca después del objeto de respuesta y se divide por un punto`


El atributo `status_code` te dice si el servidor ha respondido o si ha ocurrido un error.

Vamos a generar el código de estado de la respuesta:

`print(req.status_code)`

Resultado:

`200`

Desafortunadamente, no todas las solicitudes regresan con datos. A veces, las solicitudes devuelven errores; cada uno tiene un código especial dependiendo de su tipo. Estos son los errores más comunes:
Código de error	Nombre	implicaciones
200	OK	Todo está bien
302	Found	La página ha sido movida
400	Bad Request	Error en la sintaxis de la solicitud
404	Not Found	No se puede encontrar la página
500	Internal Server Error	Error por parte del servidor
502	Bad Gateway	Error en el intercambio de datos entre servidores
503	Server Unvailable	El servidor no puede procesar solicitudes temporalmente

`````python

import requests as rq 

URL = 'https://practicum-content.s3.us-west-1.amazonaws.com/new-markets/Data_sprint_6/web-sites/most_famous_shipwrecks_es_1.html?etag=87662559becdeb134637e5670502d6d6'

req = rq.get(URL) # guardando el objeto de respuesta como variable requerida

print(req.text) # el nombre del atributo se coloca después del objeto de respuesta y se divide por un punto
print(req.status_code) # el código de estado se muestra con el mismo formato que el texto de la respuesta
`````

## Teoría

Los analistas trabajan con grandes conjuntos de datos y recuperan información de ellos, descartando datos dudosos y buscando patrones. Aquí es donde entra en juego una poderosa herramienta para buscar cadenas: las expresiones regulares. 

Una expresión regular (regex) es una regla para buscar subcadenas (fragmentos de texto dentro de cadenas). Por ejemplo, si estamos buscando la combinación de letras "tion" en la oración "Arrived at the station in total frustration", Python lo entenderá como si estuviera buscando la subcadena `"tion"`  dentro de la cadena `'Arrived at the station in total frustration'`. 

Las expresiones regulares te permiten crear reglas complejas para que una expresión devuelva varias subcadenas. Por ejemplo, podríamos encontrar todos los naipes mencionados en Alicia en el País de las Maravillas.

**Cómo funcionan las expresiones regulares**

Para empezar a trabajar con expresiones regulares en Python, necesitamos importar el módulo `re` (expresiones regulares). Siguen dos etapas.

En la primera etapa, creamos el patrón de la expresión regular. Este es un algoritmo que describe lo que debe buscarse dentro del texto (por ejemplo, todas las letras mayúsculas).

Luego, este patrón se pasa a los métodos específicos de re. Estos métodos buscan, reemplazan y eliminan símbolos. En otras palabras, el patrón identifica qué buscar y cómo, mientras que el método define qué hacer con las coincidencias que se encuentran.

**Patrones básicos de expresiones regulares**

Ten en cuenta que, a veces, los caracteres que se utilizan en la sintaxis de expresiones regulares (estos caracteres se denominan caracteres reservados o caracteres especiales) deben utilizarse en su sentido literal. Puede que queramos, por ejemplo, hacer coincidir una subcadena que contiene un signo más literal. El uso de estos caracteres literalmente se denomina caracteres especiales de escape y se hace anteponiéndolos con una barra invertida. En otras palabras, necesitamos escribir \+ para mostrar que queremos un más literal, o \. para mostrar que queremos un punto literal. La barra invertida en sí también es un caracter especial, por lo que para una barra invertida literal, debes escribir \\.

**Crear una expresión regular**

Escribamos una expresión regular que verifique si la cadena comienza con un número y, si lo hace, coincide con el número. Iremos paso a paso.

`[0-9]`: este patrón coincide con cualquier dígito del 0 al 9 (¡solo uno a la vez!). Aplicado a la cadena `155 plus 33`, coincidirá con 1, 5, 5, 3 y 3.

`[0-9]+`: este patrón coincide con secuencias continuas de dígitos, es decir, números. Aplicado a la cadena `155 plus 33`, coincidirá con `155` y `33`.

`^[0-9]+`: ^ coincide con el comienzo de una cadena. Dado que una cadena dada solo puede tener un comienzo, este patrón siempre producirá una sola coincidencia para una cadena dada: el primer número (si la cadena comienza con un número) o ninguna coincidencia (si la cadena no comienza con un número). Aplicado a la cadena `155 plus 33`, este patrón producirá una sola coincidencia: `155`.

**Expresiones regulares en Python**

Estas son algunas de las tareas más comunes para los analistas:

* encontrar una subcadena dentro de una cadena
* dividir cadenas en subcadenas
*  reemplazar partes de una cadena con otras cadenas

Para completar estas tareas, necesitarás los siguientes métodos `re`:

1. search(patrón, cadena) busca un `patrón` en una `cadena`. Aunque `search()` recorre toda la cadena para encontrar el patrón, solo devuelve la primera subcadena que encuentra:

```
import re
string = '"General Slocum" 15 June 1904 East River human factor'
print(re.search('\w+', string))

<re.Match object; span=(1, 8), match='General'>
```

El método `search()` devuelve un objeto de tipo **match**. El parámetro span define un rango de índices que coinciden con el patrón. En nuestro caso, una comilla doble de apertura " no coincide con la regla que ignora los signos de puntuación. Es por eso que vemos los índices 1 a 8: de "G" a "l." (Recuerda que el primer índice es 0.) El parámetro match indica el valor de la propia subcadena.

El patrón `r'\w+'` coincide con cualquier subcadena que contenga uno o más caracteres o dígitos o el símbolo `"_"`. Utilizando el método `search()` determinamos que la primera palabra de la cadena coincide con el patrón. Dado que '\w+' no captura espacios, el método ha devuelto todos los caracteres antes del primer espacio.

Si necesitas recuperar la primera palabra de la cadena, el ejemplo anterior es la forma correcta de hacerlo. Primero escribes el patrón que coincide con todas las palabras en la cadena, luego lo usas en la función `search()` para obtener solo la primera coincidencia. Si bien a menudo es posible hacer coincidir la primera palabra exclusivamente con el patrón `^\w+` (recuerda que el símbolo de intercalación ^ coincide con el comienzo de la cadena), este método no es confiable. Asume que la cadena comienza con una letra, un dígito o un guión bajo, y esto no siempre es así, como lo demuestra el ejemplo anterior. El uso principal del signo de intercalación es verificar si la cadena comienza con un tipo particular de caracter o palabra. Si no es así, este tipo de búsqueda no escaneará más la cadena y simplemente no devolverá nada.

¿Cómo identificamos las palabras que se encuentran entre caracteres particulares? Intentemos hacer coincidir una frase entre comillas dobles, asumiendo que solo contiene letras y espacios.

`[A-z]` : este patrón coincide con cualquier letra, ya sea mayúscula o minúscula. Este patrón coincide con cada letra de la cadena, pero como letras separadas, no como palabras.

`[A-z ]` : (¡ten en cuenta el espacio al final!): este patrón indica que los espacios también son bienvenidos, pero aún coincide con un solo caracter a la vez, ya sea una letra o un espacio.

`[A-z ]+`: este patrón coincide con una secuencia de estos caracteres hasta que se encuentra algún otro caracter o se alcanza el final de la cadena. Entonces, todos los caracteres que no están en la lista (como dígitos o signos de puntuación) se convierten esencialmente en separadores. Si aplicamos este patrón a nuestra cadena, obtendremos tres coincidencias: General Slocum, un espacio, June (observa los espacios que rodean la palabra) y ast River human factor. `"[A-z ]+"`: este patrón coincide con una comilla doble, seguida de una secuencia de letras y espacios (1 o más, sin límite superior), seguida de otra comilla doble. Eso nos deja con "General Slocum" (contando las comillas esta vez), justo lo que estábamos buscando.

Ejemplo 4 `split(patrón, cadena)` divide una cadena en los puntos donde aparece el patrón.

```
# Ejemplo 4
import re

string = '"General Slocum" 15 June 1904 East River human factor'
print(re.split('\d+', string))

['"General Slocum" ', ' June ', ' East River human factor']
```
Como puedes ver, la cadena se ha dividido en tres partes, divididas en lugares donde el método ha encontrado el patrón definido en el argumento. Aquí, el patrón '\d+' coincide con uno o más dígitos. Siempre que split() encuentra uno o más dígitos (en este caso, 15 y 1904), divide la cadena en subcadenas. Las subcadenas que coinciden con el patrón se excluyen y no están presentes en la salida.

El parámetro **maxsplit** del método `split()` limita el número de veces que se divide la cadena.

```
# Ejemplo 5
import re

string = '"General Slocum" 15 June 1904 East River human factor'
print(re.split('\d+', string, maxsplit = 1))

['"General Slocum" ', ' June 1904 East River human factor']
```

Como puedes ver, la cadena solo se ha dividido una vez.

Al responder a tu solicitud, el servidor devuelve datos estructurados en uno de varios formatos especiales, de los cuales el más común es JSON. Parece un revoltijo de dígitos, letras, dos puntos y llaves. La pregunta es: ¿cómo se trabaja con él?

JSON significa notación de objetos de JavaScript.

Así es como se ven los datos en este formato:

```
[
  {
    "name": "General Slocum",
    "date": "June 15, 1904" 
  },
  {
    "name": "Camorta",
    "date": "May 6, 1902"
  },
  {
    "name": "Norge",
    "date": "June 28, 1904"
  }
]
```

Si JSON incluye varios elementos, se escriben entre corchetes [ ... ], al igual que en las listas. Un objeto JSON individual parece un diccionario: está delimitado por llaves y tiene pares de key : value. 

La estructura bien definida de los objetos JSON facilita tomar información, convertirla en una cadena y pasarla en una solicitud HTTP. Esa es la esencia de JSON: recopilas datos dentro de un objeto (una lista de pares key : value) y luego creas una cadena para pasarla en una solicitud. El receptor vuelve a convertir esta cadena en un objeto.

Observa que **las claves se ponen entre comillas dobles, lo cual es obligatorio en JSON**. Los valores pueden ser cadenas, números, valores booleanos, valores nulos, matrices u objetos:

{
  "name": "Camorta",
  "length": 86.9
  "decks": 3
  "date": "May 6, 1902",
  "details": {
    "place": "Bay of Bengal",
    "reason": "Force of nature"
  } 
}

JSON se abre con **llaves** (si contiene un elemento) o con **corchetes** (si tiene una lista de objetos dentro). JSON nunca contiene funciones, variables o comentarios.

**Módulo JSON**

Python tiene un módulo incorporado para trabajar con datos en formato JSON:

`import json`

Su método `json.loads()` convierte cadenas que están en formato JSON:

```
x = '{"name": "General Slocum", "date": "June 15, 1904"}'
y = json.loads(x)

print('Name : {0}, date : {1}'.format(y['name'], y['date']))

# Respuesta
Name : General Slocum, date : June 15, 1904
```

Hemos mostrado el nombre del barco y la fecha del naufragio. Pero, ¿cómo exactamente? Teníamos una cadena en formato JSON. json.loads() la ha transformado para que fuera posible abordar los valores name y date desde la función `print()`.

En este caso, el JSON contenía información sobre un solo objeto. Más a menudo, se pasa una lista completa:

```
x = '[{"name": "General Slocum", "date": "June 15, 1904"}, {"name": "Camorta", "date": "May 6, 1902"}]'
y = json.loads(x)
for i in y:
    print('Name : {0}, date : {1}'.format(i['name'], i['date']))

# Respuesta
Name : General Slocum, date : June 15, 1904
Name : Camorta, date : May 6, 1902
```

Como puedes ver, hay varios objetos y se ha llamado a un bucle para listarlos.

El método `json.dumps()`, por el contrario, convierte los datos de Python al formato JSON.

```
out = json.dumps(y)
print(out)

#Respuesta
[{"name": "General Slocum", "date": "June 15, 1904"}, {"name": "Camorta", "date": "May 6, 1902"}]
```

Hemos obtenido los datos en formato JSON.

Intentemos convertir el tipo de datos de la columna en la consulta SQL. Esto se puede hacer con la ayuda de la instrucción `CAST`:

`CAST (column_name AS data_type)`

`column_name` es el campo cuyo tipo de datos se va a convertir. `data_type` es el tipo de datos deseado. También podemos escribir lo siguiente:

`column_name :: data_type`

Veamos los tipos de datos en PostgreSQL.

Tipos de datos numéricos

**integer**: un tipo de número entero similar a int en Python. En PostgreSQL los números enteros van desde -2147483648 hasta 2147483647.

**real**: un número de coma flotante, como float en Python. La precisión de los números en el tipo real es de hasta 6 puntos decimales.

Tipos de datos de string

**'TripleTen'**: es un ejemplo de un valor de tipo string. En las instrucciones SQL va entre comillas simples.

**varchar(n)**: un string de caracteres de longitud variable donde n es el número máximo de caracteres.

**text**: un string de cualquier longitud.

Fecha y hora
Fechas y horas van entre comillas simples.

**timestamp**: fecha y hora. Similar a datetime en pandas. Este formato se usa con mayor frecuencia para almacenar eventos que ocurren varias veces al día, como registros de usuarios de sitios web.

**date**: una fecha

Lógico

**bool**: un tipo de datos lógico. Puede tener tres valores en PostgreSQL: TRUE, FALSE y NULL (desconocido).

Los resultados del análisis normalmente se presentan en un cierto orden. Para ordenar los datos por un campo, utiliza el comando ORDER BY.

Así es como se ve una instrucción con agrupación y ordenación:

```
SELECT 
    field_1, 
    field_2,
    ..., 
    field_n, 
    AGGREGATE_FUNCTION(field) AS here_you_are
FROM
    table_name
WHERE -- si es necesario
    condition
GROUP BY  
    field_1, 
    field_2, 
    ..., 
    field_n,
ORDER BY -- Enumera solamente aquellos campos por los que se van a ordenar los datos
    field_1, 
    field_2, 
    ..., 
    field_n, 
    here_you_are;
```

Cuando se utiliza `ORDER BY`, solo los campos por los que queremos ordenar los datos deben ser listados en el bloque de comandos.

Se pueden utilizar dos modificadores con el comando `ORDER BY` para ordenar los datos en columnas:

* ASC (por defecto) ordena los datos en orden ascendente.
* DESC ordena los datos en orden descendente.

```
ORDER BY

    -- ordenamos los datos en orden descendente
ORDER BY
    field_name ASC;

Así es como se ve una instrucción con una cláusula `HAVING:

La selección resultante incluirá solo aquellas filas para las que la función de agregación produce resultados que cumplen la condición indicada en el bloque `HAVING`.

`HAVING` y `WHERE` tienen mucho en común. Entonces, ¿por qué no podemos pasar todas nuestras condiciones a una de ellas? Por eso es imposible establecer parámetros de filtración que utilicen los resultados de una función de agregación con WHERE. De allí viene la necesidad de usar HAVING.

Una **subconsulta** es una consulta dentro de una consulta. Por ejemplo, un director está buscando un actor para interpretar a Hamlet. Si una subconsulta está dentro del bloque FROM, SELECT seleccionará datos de la tabla que genera la subconsulta. El nombre de la tabla se indica dentro de la consulta interna y la consulta externa se refiere a las columnas de la tabla. Las subconsultas siempre se escriben entre paréntesis:

SELECT

    -- sangra las subconsultas
    (SELECT

    FROM 
        table_name
    WHERE 
        column_name = value) AS SUBQUERY_1; 
-- no olvides darles un nombre a tus subconsultas

Esta consulta funciona, pero como viste, la subconsulta para calcular SUM(price) se ejecuta una y otra vez para cada fila. ¡Esto es ineficiente!

Afortunadamente, SQL tiene una herramienta increíblemente poderosa para estos casos: las funciones de ventana. Estas funciones te permiten realizar cálculos de agregación (como SUM, AVG, COUNT) sobre un conjunto de filas sin colapsar los resultados. Es decir, obtienes el valor agregado pero mantienes cada una de las filas originales.

**Entendiendo la ventana**

En SQL, una ventana es un conjunto de filas relacionadas sobre las que se realiza una operación. Este conjunto puede ser la tabla completa o un subconjunto específico de filas. La clave es que la función opera en esta "ventana" y devuelve un resultado para cada fila dentro de ella, agregando una nueva columna a tu resultado final.

La magia de las funciones de ventana reside en la cláusula `OVER()`. Esta es la sintaxis que le dice a SQL: "Aplica esta función de agregación no a toda la tabla, sino sobre la ventana que voy a definir".

Veamos cómo podemos resolver el problema de los libros de una manera mucho más eficiente. La solución sería así:

```
SELECT
    author_id,
    name,
    price / SUM(price) OVER () AS ratio
FROM
    books_price;
```

Fíjate en la magia de OVER(). Al dejar los paréntesis vacíos, le estamos diciendo a SQL: "Calcula el SUM(price) sobre todas las filas de la tabla (la ventana por defecto) y luego añade el resultado como una nueva columna para cada fila individual".

El resultado es el mismo, pero el motor de la base de datos lo procesa de forma mucho más inteligente. El SUM() se calcula solo una vez y se aplica a todas las filas. ¡Mucho más eficiente!

## Teoría RANK() NTILE OVER(PARTITION   GROUP BY )

Vamos a ver de dónde obtienen su nombre las funciones de ventana. Digamos que necesitamos aplicar una función de agregación a un fragmento de tabla en lugar de a toda la tabla y encontrar un valor acumulativo: por ejemplo, cómo cambia el valor acumulativo del inventario de la librería. (Al principio, la tienda ofrece solo un libro por $10; el valor acumulativo es de $10. Después adquiere un segundo libro de cinco dólares y el valor acumulativo se convierte en $15).

Para simplificar las cosas, supongamos que la tienda tiene un ejemplar de cada libro.

Entonces, ¿cómo obtendríamos el valor acumulado? Tenemos que hacer la tabla `sales_1` (la que está dentro del marco morado a continuación), `sales_2` (verde) y `sales_3` (roja) hasta llegar a `sales_i` para que cada una de ellas tenga una, dos ... i filas con ventas. Y los datos en estas filas serán acumulados: la tabla `sales_i` incluirá los datos de `sales_i-1`.

Después, hay que mostrar la suma acumulada junto con la fila de la tabla principal. Para hacerlo, tendríamos que seleccionar la tabla donde la fila en cuestión es la última fila y sumar los valores. Pero hacer tablas de este tipo lleva mucho tiempo. Aquí es donde viene bien utilizar las tablas virtuales: se ejecutan a través de los datos a medida que se ejecuta la consulta. Estas tablas virtuales son, de hecho, ventanas.

image

Para que una ventana funcione correctamente es importante indicar los parámetros de ordenación para que el SGBD entienda en qué orden están los datos. 

Vamos a obtener nuestro valor acumulado:

```

    author_id,
    name,
    SUM(price) OVER (ORDER BY author_id ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW)
FROM
    books_price;

Eliminar simplemente las filas NULL no siempre es la mejor opción. A veces querremos reemplazarlas con un cierto valor (digamos, la media) y podemos hacer que esto suceda en la consulta en sí. Para eso tenemos el operador `CASE` ("en casos cuando"). Es como  if-elif-else en Python:

```
CASE WHEN condition_1 THEN
    result_1
WHEN condition_2 THEN
    result_2
WHEN condition_3 THEN
    result_3
ELSE
    result_4
END;
```

La condición va después del operador WHEN. Si una fila de la tabla cumple esta condición, el código devuelve el resultado indicado en THEN. Si no, la misma fila se prueba con la siguiente condición. Si la fila no cumple con ninguna de las condiciones establecidas en WHEN, el código devuelve el valor indicado después de ELSE. Luego, la declaración CASE se cierra con el operador END.

¿Cómo funciona esto en SQL? Vamos a ver un ejemplo:

```
SELECT
    name,
    CASE WHEN publisher_id IS NULL THEN -1 -- si no se especifica la editorial, 
-- un publisher_id de -1
    ELSE publisher_id END AS publisher_id_full
FROM
    books;
```

## JOIN, INNER JOIN

Es raro que todos los datos se almacenen en una tabla. Los analistas normalmente necesitan fusionar tablas; por eso tenemos el operador JOIN.

Hay dos formas de unir tablas: INNER y OUTER JOIN.

![alt text](imgs/015.jpg)

`INNER JOIN` devuelve solo aquellas filas que tienen valores coincidentes de una tabla a otra (la intersección de las tablas). `OUTER JOIN` recupera todos los datos de una tabla y agrega datos de la otra cuando hay filas coincidentes. Hay dos tipos de `OUTER JOIN`, left (izquierda ) y right (derecha):

![alt text](imgs/016.png)

**INNER JOIN**

INNER JOIN selecciona solo los datos para los que se cumple la condición de unión. El orden en que se unen las tablas no afecta el resultado final.

Aquí tienes un ejemplo de consulta con INNER JOIN:

```
SELECT --enumerando solo los campos que se necesitan
    TABLE_1.field_1 AS field_1,
    TABLE_1.field_2 AS field_2,
    ...
    TABLE_2.field_n AS field_n
FROM
    TABLE_1
    INNER JOIN TABLE_2 ON TABLE_2.field_1 = TABLE_1.field_2;
```

Vamos a echar un vistazo más de cerca a la sintaxis.

* `INNER JOIN` es el nombre del método de unión. Luego viene el nombre de la tabla que se unirá a la tabla del bloque `FROM`.
* `ON` precede a la condición de unión: `TABLE_2.field_1 = TABLE_1.field_2`. Esto significa que solo se unirán las filas de la tabla que cumplan esta condición. En nuestro caso, la condición es que field_1 de la segunda tabla coincida con field_2 de la primera.

Dado que los campos en diferentes tablas pueden tener los mismos nombres, se hace referencia a ellos tanto por el nombre de la tabla como por el nombre del campo. Primero va el nombre de la tabla, después el campo: `TABLE_1.field_1`.
