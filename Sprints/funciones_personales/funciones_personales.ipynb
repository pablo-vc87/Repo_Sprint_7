## Muestra los valores NaN de un dataframe y/o columna(s)

```python
# CREAMOS UNA FUNCIÓN QUE CALCULA Y MUESTRA LOS NAN DE UN DataFrame

def mostrar_nan(df, columna=None, mostrar=False, num_filas=10):
    """
    Imprime la cantidad de valores Nan por columna del DataFrame
    Muestra las filas completas que contienen valores NaN
    
    Parámetros:
    df: DataFrame a analizar
    columna: nombre de columna específica (opcional)
    num_filas: cuántas filas mostrar (por defecto 10)
    """
    name = str(df)
    lista= df.columns

    #Para una sola columna:
    if columna:  
        filas_nan = df[df[columna].isna()] #crea DF de puros valores NaN
        if mostrar:
            print(f"\nFilas con NaN en '{columna}': {len(filas_nan)} encontradas") # imprime nombre de la columna y cantidad de NaN
        if not filas_nan.empty: # Si el DF filas_nan tiene algo adentro
            if mostrar:
                print("\nPrimeras", min(num_filas, len(filas_nan)), "filas completas conteniendo NaN para: ", "'",columna,"'")
                print(filas_nan.head(num_filas))
                print('-'*50) #separador visual
    else:
        filas_nan =df[df.isna().any(axis=1)]
        if mostrar:
            # 1. Mostrar total de filas con NaN
            print(f"\nTotal de filas con NaN: {len(filas_nan)}")
    
            # 2. Mostrar NaN por cada columna del DataFrame ORIGINAL
            print(f"\nCantidad de NaN por columna:")
            for column in df.columns:  # df.columns, no filas_nan.columns
                nan_count = df[column].isna().sum()  # Contar NaN en cada columna
                print(f"  {column}: {nan_count}")
    
            # 3. Mostrar las primeras filas con NaN
            if not filas_nan.empty:
                print(f"\nPrimeras {min(num_filas, len(filas_nan))} filas con NaN:")
                print(filas_nan.head(num_filas))
                print('-'*50)
    return filas_nan
```

## Muestra filas duplicadas de un dataframe y/o columnas

```python
def mostrar_dupli(df, columna=None, num_filas=10):
    """
    Imprime la cantidad de filas duplicadas del DataFrame
    Muestra las filas completas que están duplicadas
    
    Parámetros:
    df: DataFrame a analizar
    columna: nombre de columna específica para buscar duplicados (opcional)
    num_filas: cuántas filas mostrar (por defecto 10)
    """
    
    print(f"\nEl dataframe tiene {df.duplicated().sum()} filas completamente duplicadas")
    print('-'*50)
    # Para una sola columna:
    if columna:  
        filas_duplicadas = df[df.duplicated(subset=[columna], keep=False)]
        print(f"\nFilas con valores duplicados en '{columna}': {len(filas_duplicadas)} encontradas")
        print('-'*50)
        if not filas_duplicadas.empty:
            print(f"\nPrimeras {min(num_filas, len(filas_duplicadas))} filas completas con duplicados en: '{columna}'")
            print(filas_duplicadas.head(num_filas))
            print('-'*50)
    else:        
        # Para duplicados completos (todas las columnas)
        filas_duplicadas = df[df.duplicated(keep=False)]
        print(f"\nFilas completamente duplicadas: {len(filas_duplicadas)} encontradas")
        print('-'*50)
        if not filas_duplicadas.empty:
            print(f"\nPrimeras {min(num_filas, len(filas_duplicadas))} filas completamente duplicadas:")
            print(filas_duplicadas.sort_values('user_id').head(num_filas))
            print('-'*50)
    return filas_duplicadas
```

## Estadísticos explícitos de una boxplot

```python
def boxplot_stats(datos, nombre='Datos', atipicos=False):
    '''
    Requiere importar numpy as np.
    Función que recibe un array y el nombre del array (Datos por defecto)
    y puede devolver un diccionario con los estadísticos y valores atipicos
    si el tercer argumento es True.
    Arroja información completa de los puntos más importantes de
    un diagrama de caja incluyendo los valores atípicos.
    '''
    ## Dato límite del primer cuartil
    q1 = np.percentile(datos, 25)
    ## Dato límite del tercer cuartil
    q3 = np.percentile(datos, 75)
    ## Dato de cuánto mide el espacio intercuartil
    iqr = q3 - q1
    
    # Límites teóricos de bigotes
    ## valor del primer cuartil menos 1.5 veces lo que mide el espacio intercuartil
    limite_inf = q1 - 1.5 * iqr
    ## valor del tercer cuartil mas 1.5 veces lo que mide el espacio intercuartil
    limite_sup = q3 + 1.5 * iqr
    
    # Valores reales de los bigotes
    ## El mínimo de los valores que superen el límite inferior
    min_bigote = np.min(datos[datos >= limite_inf])
    ## El máximo de los valores que esten debajo del límite superior
    max_bigote = np.max(datos[datos <= limite_sup])

    # Identificar los valores atípicos
    ## Todos los datos por debajo del límite inferior
    atipicos_inf = datos[datos < limite_inf]
    ## Todos los datos por encima del límite superior
    atipicos_sup = datos[datos > limite_sup]
    ## Total en número de los valores atípicos
    atipicos_totales = len(atipicos_inf) + len(atipicos_sup)
    ## Procentaje de valores atípicos respecto al total de datos
    porcentaje_atipicos= (atipicos_totales/len(datos)) * 100

    # Imprime los datos del boxplot del arreglo
    print(f"=== {nombre.upper()}===")
    print(f"Mínimo absoluto: {np.min(datos):.2f}")
    print(f"Mínimo del bigote: {min_bigote:.2f}")
    print(f"Q1: {q1:.2f}")
    print(f"Mediana: {np.percentile(datos, 50):.2f}")
    print(f"Q3: {q3:.2f}")
    print(f"Máximo del bigote: {max_bigote:.2f}")
    print(f"Máximo absoluto: {np.max(datos):.2f}")
    print(f"\n--- VALORES ATÍPICOS ---")
    print(f"Atípicos inferiores: {len(atipicos_inf)} valores.")
    print(f"Atípicos superiores: {len(atipicos_sup)} valores.")
    print(f"Atípicos totales: {atipicos_totales} ({porcentaje_atipicos:.1f}% del total.)")
    print(f"Datos normales: {len(datos) - atipicos_totales} ({100 - porcentaje_atipicos:.1f}% del total.)")
    if atipicos:
        return{'atipicos_inf':atipicos_inf,
               'atipicos_sup':atipicos_sup,
               'stats': {'q1': q1, 'q3': q3, 'mediana' : np.percentile(datos, 50)}
              }
```

```python
def crear_boxplots_mensuales(df, columnas_base, nombre_metrica, filtrar_ceros=True, plan_surf='surf', plan_ultimate='ultimate'):
    """
    Crea boxplots mensuales en facets para cualquier métrica
    
    Parámetros:
    df: DataFrame con los datos
    columnas_base: string base para las columnas (ej: 'mins', 'msgs', 'gb')
    nombre_metrica: string para títulos (ej: 'Minutos', 'Mensajes', 'GB')
    filtrar_ceros: bool, si filtrar valores = 0
    plan_surf: nombre del plan surf en los datos
    plan_ultimate: nombre del plan ultimate en los datos
    """
    
    fig, axes = plt.subplots(3, 4, figsize=(15, 10))
    axes = axes.flatten()
    
    # Título principal
    titulo = f'Distribución de {nombre_metrica} Mensuales por Plan - Boxplots por Mes'
    if filtrar_ceros:
        titulo += ' (Usuarios Activos)'
    fig.suptitle(titulo, fontsize=16, fontweight='bold')
    
    for i, mes in enumerate(range(1, 13)):
        columna_mes = f'{columnas_base}_{mes}'
        
        # Datos del mes específico
        surf_data = df[df['plan_name'] == plan_surf][columna_mes].dropna()
        ultimate_data = df[df['plan_name'] == plan_ultimate][columna_mes].dropna()
        
        # Filtrar valores mayores a 0 si se especifica
        if filtrar_ceros:
            surf_data = surf_data[surf_data > 0]
            ultimate_data = ultimate_data[ultimate_data > 0]
        
        # Crear boxplot para el mes
        datos_para_boxplot = []
        etiquetas = []
        
        if len(surf_data) > 0:
            datos_para_boxplot.append(surf_data)
            etiquetas.append('Surf')
        
        if len(ultimate_data) > 0:
            datos_para_boxplot.append(ultimate_data)
            etiquetas.append('Ultimate')
        
        if datos_para_boxplot:
            box_plot = axes[i].boxplot(datos_para_boxplot, 
                                     labels=etiquetas,
                                     patch_artist=True)
            
            # Colorear las cajas
            colors = ['lime', 'hotpink']
            for patch, color in zip(box_plot['boxes'], colors[:len(box_plot['boxes'])]):
                patch.set_facecolor(color)
                patch.set_alpha(0.7)
        
        axes[i].set_title(f'Mes {mes}')
        axes[i].set_ylabel(nombre_metrica)
        axes[i].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
#===================================
def crear_histogramas_mensuales(df, tipo_dato='mins', mostrar_solo_activos=True, bins=20, figsize=(15, 10)):
    """
    Crea histogramas mensuales sin usar flatten(), procesando mes por mes
    """
    # Validar tipo de dato
    if tipo_dato not in ['mins', 'msgs', 'gb']:
        raise ValueError("tipo_dato debe ser 'mins', 'msgs' o 'gb'")

    # Definir unidades

    unidades = {'mins': 'Minutos', 'msgs': 'Mensajes', 'gb': 'GB'}
    unidad = unidades[tipo_dato]

    # Crear figura con subplots
    fig, axes = plt.subplots(3, 4, figsize=figsize)
    axes = axes.flatten()

    # Título principal
    titulo = f'Distribución de {unidad} Mensuales por Plan - Año 2018'
    if mostrar_solo_activos:
        titulo += ' (Solo Usuarios Activos)'
    fig.suptitle(titulo, fontsize=16, fontweight='bold')

    # Procesar cada mes por separado (SIN FLATTEN)
    for i, mes in enumerate(range(1, 13)):
        columna_mes = f'{tipo_dato}_{mes}'

        # Filtrar datos específicos para cada mes
        #surf_data_mes = dfdf['plan_name']=='surf'.dropna()
        #ultim_data_mes = dfdf['plan_name']=='ultimate'.dropna()
        surf_data_mes = df[df['plan_name']=='surf'][columna_mes].dropna()
        ultim_data_mes = df[df['plan_name']=='ultimate'][columna_mes].dropna()

        # Filtrar solo valores mayores a 0 si se especifica
        if mostrar_solo_activos:
            surf_data_mes = surf_data_mes[surf_data_mes > 0]
            ultim_data_mes = ultim_data_mes[ultim_data_mes > 0]

        # Crear histograma para este mes específico
        axes[i].hist(surf_data_mes, bins=bins, alpha=0.7, color='lime', label='Surf')
        axes[i].hist(ultim_data_mes, bins=bins, alpha=0.7, color='hotpink', label='Ultimate')
        axes[i].set_title(f'Mes {mes}')
        axes[i].set_xlabel(unidad)
        axes[i].set_ylabel('Frecuencia')
        axes[i].legend(fontsize=8)

    plt.tight_layout()
    plt.show()

#===============================================================
def analizar_estadisticas_mensuales(df, tipo_dato='mins', mostrar_solo_activos=True):
    """
    Analiza y grafica la evolución temporal de medias y varianzas mensuales por plan.

    Parámetros:
    df: DataFrame con los datos (df_ready_for_stats_copy recomendado)
    tipodato: str - Tipo de dato a analizar ('mins', 'msgs', 'gb')
    mostrar_solo_activos: bool - Si True, filtra solo usuarios con valores > 0

    Retorna:
    dict: Diccionario con las estadísticas calculadas
    """

    # Validar tipo de dato
    if tipo_dato not in ['mins', 'msgs', 'gb']:
        raise ValueError("tipo_dato debe ser 'mins', 'msgs' o 'gb'")

    # Definir unidades y etiquetas
    unidades = {
        'mins': 'Minutos',
        'msgs': 'Mensajes', 
        'gb': 'GB'
    }

    unidad = unidades[tipo_dato]
    meses = list(range(1, 13))

    # Inicializar listas para almacenar estadísticas
    medias_surf = []
    varianzas_surf = []
    medias_ultimate = []
    varianzas_ultimate = []

    # Calcular estadísticas para cada mes
    for mes in meses:
        columna_mes = f'{tipo_dato}_{mes}'

        # Obtener datos del mes para cada plan
        surf_mes = df[df['plan_name']=='surf'][columna_mes].dropna()
        ultimate_mes = df[df['plan_name']=='ultimate'][columna_mes].dropna()

        # Filtrar solo usuarios activos si se especifica
        if mostrar_solo_activos:
            surf_mes = surf_mes[surf_mes > 0]
            ultimate_mes = ultimate_mes[ultimate_mes > 0]

        # Calcular y almacenar estadísticas
        medias_surf.append(surf_mes.mean())
        varianzas_surf.append(surf_mes.var())
        medias_ultimate.append(ultimate_mes.mean())
        varianzas_ultimate.append(ultimate_mes.var())

    # Crear los gráficos
    fig, axes = plt.subplots(2, 1, figsize=(12, 10))

    # Título principal
    titulo_base = f'Evolución de Estadísticas de {unidad} por Mes'
    if mostrar_solo_activos:
        titulo_base += ' (Solo Usuarios Activos)'

    # Gráfico de evolución de medias
    axes[0].plot(meses, medias_surf, marker='o', color='lime', label='Surf', linewidth=2)
    axes[0].plot(meses, medias_ultimate, marker='s', color='hotpink', label='Ultimate', linewidth=2)
    axes[0].set_title(f'Evolución de la Media de {unidad} por Mes', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Mes')
    axes[0].set_ylabel(f'Media de {unidad}')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Gráfico de evolución de medias
    axes[1].plot(meses, varianzas_surf, marker='o', color='lime', label='Surf', linewidth=2)
    axes[1].plot(meses, varianzas_ultimate, marker='s', color='hotpink', label='Ultimate', linewidth=2)
    axes[1].set_title(f'Evolución de la Varianza de {unidad} por Mes', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Mes')
    axes[1].set_ylabel(f'Varianza de {unidad}')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def crear_histogramas_de_ingresos_mensuales(df, tipo_dato='total',
                                            mostrar_solo_activos=True,
                                            bins=20, figsize=(15, 10),
                                            mes_inicio = 1,
                                            mes_final =12
                                           ):
    """
    Crea histogramas mensuales sin usar flatten(), procesando mes por mes
    """
    # Validar tipo de dato
    if tipo_dato not in ['base', 
                         'extra_mins',
                         'extra_msgs',
                         'extra_gb',
                         'total'
                        ]:
        raise ValueError("tipo_dato debe ser 'base', 'extra_msgs', 'extra_mins', 'extra_gb' o 'total'"
                         )

    # Definir unidades
    unidades = {'base' : 'Ingreso Base',
                 'extra_mins' : 'Ingreso Extra Minutos',
                 'extra_msgs' : 'Ingreso Extra Mensajes',
                 'extra_gb' : 'Ingreso Extra gb',
                 'total' : 'ingreso Total'
                 }

    unidad = unidades[tipo_dato]
    meses = list(range(mes_inicio, mes_final + 1))

    # Para ingresos base, usar bins específicos que capturen $20 y $70
    if tipo_dato == 'base':
        # Para ingresos base, usar bins específicos que capturen $20 y $70
        bins_especiales = [19.5, 20.5, 69.5, 70.5]  # Bins centrados en los valores exactos
        
        # Crear figura con subplots
        fig, axes = plt.subplots(3, 4, figsize=figsize)
        axes = axes.flatten()
        
        for i, mes in enumerate(range(mes_inicio, mes_final +1)):
            columna_mes = f'ingreso_{tipo_dato}_{mes}'
            
            surf_data_mes = df[df['plan_name']=='surf'][columna_mes].dropna()
            ultim_data_mes = df[df['plan_name']=='ultimate'][columna_mes].dropna()
            
            # Para ingresos base, contar frecuencias exactas
            surf_20 = (surf_data_mes == 20).sum()
            ultim_70 = (ultim_data_mes == 70).sum()
            
            # Crear gráfico de barras en lugar de histograma
            axes[i].bar(['Surf ($20)', 'Ultimate ($70)'], 
                       [surf_20, ultim_70], 
                       color=['lime', 'hotpink'], 
                       alpha=0.7)
            
            axes[i].set_title(f'Mes {mes}')
            axes[i].set_ylabel('Cantidad de usuarios')
            
        plt.tight_layout()
        plt.show()
    else:
        # Crear figura con subplots
        fig, axes = plt.subplots(3, 4, figsize=figsize)
        axes = axes.flatten()
        
        # Título principal
        titulo = f'Distribución de {unidad} Mensuales por Plan - Año 2018'
        if mostrar_solo_activos:
            titulo += ' (Solo Usuarios Activos)'
        fig.suptitle(titulo, fontsize=16, fontweight='bold')
    
        # Procesar cada mes por separado (SIN FLATTEN)
        for i, mes in enumerate(range(mes_inicio, mes_final +1)):
            columna_mes = f'ingreso_{tipo_dato}_{mes}'
    
            # Filtrar datos específicos para cada mes
            #surf_data_mes = dfdf['plan_name']=='surf'.dropna()
            #ultim_data_mes = dfdf['plan_name']=='ultimate'.dropna()
            surf_data_mes = df[df['plan_name']=='surf'][columna_mes].dropna()
            ultim_data_mes = df[df['plan_name']=='ultimate'][columna_mes].dropna()
    
         # Para ingresos base, contar frecuencias exactas
            surf_20 = (surf_data_mes == 20).sum()
            ultim_70 = (ultim_data_mes == 70).sum()
            
            # Filtrar solo valores mayores a 0 si se especifica
            if mostrar_solo_activos:
                surf_data_mes = surf_data_mes[surf_data_mes > 0]
                ultim_data_mes = ultim_data_mes[ultim_data_mes > 0]
    
            # Crear histograma para este mes específico
            axes[i].hist(surf_data_mes, bins=bins, alpha=0.7, color='lime', label='Surf')
            axes[i].hist(ultim_data_mes, bins=bins, alpha=0.7, color='hotpink', label='Ultimate')
            axes[i].set_title(f'Mes {mes}')
            axes[i].set_xlabel(unidad)
            axes[i].set_ylabel('Frecuencia')
            axes[i].legend(fontsize=8)

    plt.tight_layout()
    plt.show()
#==============================================================================
def crear_histogramas_de_ingresos_por_meses(df, tipo_dato='total',
                                            mostrar_solo_activos=True,
                                            bins=20, figsize=(15, 10),
                                            mes_inicio = 1,
                                            mes_final =12
                                           ):
    """
    Crea histogramas mensuales sin usar flatten(), procesando mes por mes
    """
    # Validar tipo de dato
    if tipo_dato not in ['base', 
                         'extra_mins',
                         'extra_msgs',
                         'extra_gb',
                         'total'
                        ]:
        raise ValueError("tipo_dato debe ser 'base', 'extra_msgs', 'extra_mins', 'extra_gb' o 'total'"
                         )

    # Definir unidades
    unidades = {'base' : 'Ingreso Base',
                 'extra_mins' : 'Ingreso Extra Minutos',
                 'extra_msgs' : 'Ingreso Extra Mensajes',
                 'extra_gb' : 'Ingreso Extra gb',
                 'total' : 'ingreso Total'
                 }

    unidad = unidades[tipo_dato]
    meses = list(range(mes_inicio, mes_final + 1))

    
    # Crear figura con subplots
    fig, axes = plt.subplots(3, 4, figsize=figsize)
    axes = axes.flatten()
    
    # Título principal
    titulo = f'Distribución de Ingreso por ciudades Mensuales- Oct-Dic'
    if mostrar_solo_activos:
        titulo += ' (Solo Usuarios Activos)'
    fig.suptitle(titulo, fontsize=16, fontweight='bold')

    # Procesar cada mes por separado (SIN FLATTEN)
    for i, mes in enumerate(range(mes_inicio, mes_final +1)):
        columna_mes = f'ingreso_{tipo_dato}_{mes}'

        # Filtrar datos específicos para cada mes
        #surf_data_mes = dfdf['plan_name']=='surf'.dropna()
        #ultim_data_mes = dfdf['plan_name']=='ultimate'.dropna()
        
       
        ny_nj_data_mes= df[df['city'].str.contains('NY-NJ')][columna_mes].dropna()
        other_cities_data_mes = df[~df['city'].str.contains('NY-NJ')][columna_mes].dropna()


        
        # Filtrar solo valores mayores a 0 si se especifica
        if mostrar_solo_activos:
            ny_nj_data_mes = ny_nj_data_mes[ny_nj_data_mes > 0]
            other_cities_data_mes = other_cities_data_mes[other_cities_data_mes > 0]

        # Crear histograma para este mes específico
        axes[i].hist(other_cities_data_mes, bins=bins, alpha=0.5, color='lime', label='NY-NJ')
        axes[i].hist(ny_nj_data_mes, bins=bins, alpha=0.5, color='hotpink', label='Other cities')
        axes[i].set_title(f'Mes {mes}')
        axes[i].set_xlabel(unidad)
        axes[i].set_ylabel('Frecuencia')
        axes[i].legend(fontsize=8)

    plt.tight_layout()
    plt.show()
#=========================================
def analizar_estadisticas_de_ingresos_por_meses(df, tipo_dato='total', 
                                                mostrar_solo_activos=True,
                                                mes_inicio= 1,
                                                mes_final= 12
                                               ):
    """
    Analiza y grafica la evolución temporal de medias,
    medianas y desviación estandar mensuales por plan.

    Parámetros:
    df: DataFrame con los datos (df_ready_for_stats_copy recomendado)
    tipodato: str - Tipo de dato a analizar ('mins', 'msgs', 'gb')
    mostrar_solo_activos: bool - Si True, filtra solo usuarios con valores > 0

    Retorna:
    dict: Diccionario con las estadísticas calculadas
    """

    # Validar tipo de dato
    if tipo_dato not in ['base', 
                         'extra_mins',
                         'extra_msgs',
                         'extra_gb',
                         'total'
                        ]:
        raise ValueError("tipo_dato debe ser 'base', 'extra_msgs', 'extra_mins', 'extra_gb' o 'total'"
                         )

    # Definir unidades y etiquetas
    unidades = {
        'base' : 'Ingreso Base ($)',
        'extra_mins' : 'Ingreso Extra Minutos',
        'extra_msgs' : 'Ingreso Extra Mensajes',
        'extra_gb' : 'Ingreso Extra gb',
        'total' : 'ingreso Total'
    }

    unidad = unidades[tipo_dato]
    meses = list(range(mes_inicio, mes_final +1))

    # Inicializar listas para almacenar estadísticas
    medias_surf = []
    varianzas_surf = []
    desv_std_surf = []
    medianas_surf= []
    medias_ultimate = []
    varianzas_ultimate = []
    desv_std_ultimate = []
    medianas_ultimate= []

    # Calcular estadísticas para cada mes
    for mes in meses:
        columna_mes = f'ingreso_{tipo_dato}_{mes}'

        # Obtener datos del mes para cada plan
        ny_nj_data_mes= df[df['city'].str.contains('NY-NJ')][columna_mes].dropna()
        other_cities_data_mes = df[~df['city'].str.contains('NY-NJ')][columna_mes].dropna()

        # Filtrar solo usuarios activos si se especifica
        if mostrar_solo_activos:
            ny_nj_data_mes = ny_nj_data_mes[ny_nj_data_mes > 0]
            other_cities_data_mes = other_cities_data_mes[other_cities_data_mes > 0]

        if len(ny_nj_data_mes) >=2 :
            medias_surf.append(ny_nj_data_mes.mean())
            medianas_surf.append(ny_nj_data_mes.median())
            varianzas_surf.append(ny_nj_data_mes.var())
            desv_std_surf.append(ny_nj_data_mes.std())
        else:
            medias_surf.append(ny_nj_data_mes.mean() if len(ny_nj_data_mes) > 0 else np.nan)
            medianas_surf.append(ny_nj_data_mes.median() if len(ny_nj_data_mes) > 0 else np.nan)
            varianzas_surf.append(np.nan)  # No se puede calcular con < 2 valores
            desv_std_surf.append(np.nan)
            
        # Mismo tratamiento para Ultimate
        if len(other_cities_data_mes) >= 2:
            medias_ultimate.append(other_cities_data_mes.mean())
            medianas_ultimate.append(other_cities_data_mes.median())
            varianzas_ultimate.append(other_cities_data_mes.var())
            desv_std_ultimate.append(other_cities_data_mes.std())
        else:
            medias_ultimate.append(other_cities_data_mes.mean() if len(other_cities_data_mes) > 0 else np.nan)
            medianas_ultimate.append(other_cities_data_mes.median() if len(other_cities_data_mes) > 0 else np.nan)
            varianzas_ultimate.append(np.nan)
            desv_std_ultimate.append(np.nan)
            

            

        # Calcular y almacenar estadísticas
        #medias_surf.append(surf_mes.mean())
        #varianzas_surf.append(surf_mes.var())
        #desv_std_surf.append(surf_mes.std())
        #medias_ultimate.append(ultimate_mes.mean())
        #varianzas_ultimate.append(ultimate_mes.var())
        #desv_std_ultimate.append(ultimate_mes.std())

    # Crear los gráficos
    fig, axes = plt.subplots(3, 1, figsize=(12, 10))

    # Título principal
    titulo_base = f'Evolución de Estadísticas de {unidad} por Mes'
    if mostrar_solo_activos:
        titulo_base += ' (Solo Usuarios Activos)'
    # Aplicar el título a la figura DESPUÉS de crearlo
    fig.suptitle(titulo_base, fontsize=16, fontweight='bold')  # ← AGREGAR ESTA LÍNEA

    # Gráfico de evolución de medias
    axes[0].plot(meses, medias_surf, marker='o', color='lime', label='NY-NJ', linewidth=2)
    axes[0].plot(meses, medias_ultimate, marker='s', color='hotpink', label='Other cities', linewidth=2)
    axes[0].set_title(f'Evolución de la Media de {unidad} de Mes {mes_inicio} a Mes {mes_final}', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Mes')
    axes[0].set_ylabel(f'Media de {unidad}')
    axes[0].legend()
    axes[0].grid(True, alpha=0.5)

    # Gráfico de evolución de varianzas
    axes[1].plot(meses, varianzas_surf, marker='o', color='lime', label='NY-NJ', linewidth=2)
    axes[1].plot(meses, varianzas_ultimate, marker='s', color='hotpink', label='Other cities', linewidth=2)
    axes[1].set_title(f'Evolución de la Varianza de {unidad} de Mes {mes_inicio} a Mes {mes_final}', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Mes')
    axes[1].set_ylabel(f'Varianza de {unidad}')
    axes[1].legend()
    axes[1].grid(True, alpha=0.5)

     # Gráfico de evolución de desviaciones std
    axes[2].plot(meses, desv_std_surf, marker='o', color='lime', label='NY-NJ', linewidth=2)
    axes[2].plot(meses, desv_std_ultimate, marker='s', color='hotpink', label='Other cities', linewidth=2)
    axes[2].set_title(f'Evolución de la Desviacion Estándar de {unidad} de Mes {mes_inicio} a Mes {mes_final}', fontsize=14, fontweight='bold')
    axes[2].set_xlabel('Mes')
    axes[2].set_ylabel(f'Desviación estándar de {unidad}')
    axes[2].legend()
    axes[2].grid(True, alpha=0.5)
    
    
    plt.tight_layout()
    plt.show()
```

## muestra los análisis de columna categórica y columna numérica de un df

```python
def analisis_completo_categoria(df, columna_categoria, columna_numerica='total_sales', top_n=3, descripcion=False):
    """
    Realiza análisis completo para una columna categórica
    
    Parámetros:
    df: DataFrame
    columna_categoria: nombre de la columna categórica (ej: 'genre', 'platform')
    columna_numerica: columna numérica para analizar (por defecto 'total_sales')
    top_n: número de elementos top a mostrar por categoría
    
    Retorna: diccionario con todos los análisis
    """
    if descripcion:     
        print(f"{'='*60}")
        print(f"ANÁLISIS COMPLETO PARA: {columna_categoria.upper()} en {columna_numerica}")
        print(f"{'='*60}")
    
    resultados = {}
    
    # 1. VENTAS TOTALES POR CATEGORÍA
    if descripcion:
        print(f"\n1. VENTAS TOTALES POR {columna_categoria.upper()} (para {columna_numerica}):")
    ventas_totales = df.groupby(columna_categoria)[columna_numerica].sum().sort_values(ascending=False)
    resultados['ventas_totales'] = {
        'categorias': ventas_totales.index.tolist(),
        'valores': [round(i, 2) for i in ventas_totales.values]
    }
    if descripcion:
        print(f"Categorías: {resultados['ventas_totales']['categorias']}")
        print(f"Ventas: {resultados['ventas_totales']['valores']}")
    
    # 2. VENTAS PROMEDIO POR CATEGORÍA
    if descripcion:
        print(f"\n2. VENTAS PROMEDIO POR {columna_categoria.upper()}: (para {columna_numerica})")
    ventas_promedio = df.groupby(columna_categoria)[columna_numerica].mean().sort_values(ascending=False)
    resultados['ventas_promedio'] = {
        'categorias': ventas_promedio.index.tolist(),
        'valores': [round(i, 2) for i in ventas_promedio.values]
    }
    if descripcion:
        print(f"Categorías: {resultados['ventas_promedio']['categorias']}")
        print(f"Promedios: {resultados['ventas_promedio']['valores']}")
    
    # 3. CANTIDAD DE JUEGOS POR CATEGORÍA
    if descripcion:
        print(f"\n3. CANTIDAD DE JUEGOS POR {columna_categoria.upper()}: (para {columna_numerica})")
    cantidad_juegos = df.groupby(columna_categoria)['name'].count().sort_values(ascending=False)
    resultados['cantidad_juegos'] = {
        'categorias': cantidad_juegos.index.tolist(),
        'valores': cantidad_juegos.values.tolist()
    }
    if descripcion:
        print(f"Categorías: {resultados['cantidad_juegos']['categorias']}")
        print(f"Cantidades: {resultados['cantidad_juegos']['valores']}")

    # 4. PARTICIPACIÓN DE MERCADO
    if descripcion:
        print(f"\n4. PARTICIPACIÓN DE MERCADO POR {columna_categoria.upper()} (para {columna_numerica}):")
    total_mercado = df[columna_numerica].sum()
    participacion_mercado = (ventas_totales / total_mercado * 100).sort_values(ascending=False)
    resultados['participacion_mercado'] = {
        'categorias': participacion_mercado.index.tolist(),
        'valores': [round(i, 2) for i in participacion_mercado.values]
    }
    if descripcion:
        print(f"Categorías: {resultados['participacion_mercado']['categorias']}")
        print(f"Participación (%): {resultados['participacion_mercado']['valores']}")
    
    # 5. TOP N JUEGOS POR CATEGORÍA (NUEVO)
    if descripcion:
        print(f"\n5. TOP {top_n} JUEGOS POR {columna_categoria.upper()} (para {columna_numerica}):")
    top_juegos_por_categoria = {}
    for categoria in df[columna_categoria].unique():
        juegos_categoria = df[df[columna_categoria] == categoria].nlargest(top_n, columna_numerica)
        top_juegos_por_categoria[categoria] = {
            'nombres': juegos_categoria['name'].tolist(),
            'ventas': [round(i, 2) for i in juegos_categoria[columna_numerica].tolist()]
        }
        if descripcion:
            print(f"\n{categoria}:")
            for i, (nombre, venta) in enumerate(zip(top_juegos_por_categoria[categoria]['nombres'], 
                                                   top_juegos_por_categoria[categoria]['ventas']), 1):
                print(f"  {i}. {nombre}: {venta}")
    
    resultados['top_juegos_por_categoria'] = top_juegos_por_categoria
    
    return resultados

#============================================================================================
#===============================================================================
def graficas_analisis_completo(datos_analisis, 
                               titulo_principal,
                               columna_categoria, 
                               columna_numerica='total_sales', 
                               figsize=(20, 15),
                               max_categorias_vt=None, 
                               max_categorias_vp=None,
                               max_categorias_cj=None, 
                               max_categorias_pm=None,
                               max_categorias_grafico5=None,
                               max_categorias_grafico6=None):
    """
    Crea gráficos para todos los análisis de la función analisis_completo_categoria()
    
    Parámetros:
    datos_analisis: diccionario retornado por analisis_completo_categoria()
    titulo_principal: título principal para toda la figura
    columna_categoria: nombre de la columna categórica analizada
    columna_numerica: columna numérica analizada
    figsize: tamaño de la figura completa
    max_categorias_vt: máximo de categorías para ventas totales (None = todas)
    max_categorias_vp: máximo de categorías para ventas promedio (None = todas)
    max_categorias_cj: máximo de categorías para cantidad juegos (None = todas)
    max_categorias_pm: máximo de categorías para participación mercado (None = todas)
    max_categorias_grafico5: máximo de categorías para Top 3 juegos (None = 4)
    max_categorias_grafico6: máximo de categorías para Top vs Bottom (None = 6, 3 vs 3)
    """
    
    # Función auxiliar para generar colores dinámicamente
    def generar_colores(n_categorias):
        if n_categorias <= 10:
            colores_base = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', 
                           '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
            return colores_base[:n_categorias]
        else: 
            import matplotlib.cm as cm
            cmap = cm.get_cmap('tab20') 
            return [cmap(i/n_categorias) for i in range(n_categorias)]
    
    # Función auxiliar para preparar datos con "Otros"
    def preparar_datos_con_otros(categorias, valores, max_mostrar):
        if max_mostrar is None or len(categorias) <= max_mostrar:
            return categorias, valores
        else:
            categorias_mostrar = categorias[:max_mostrar] + ['Otros']
            valores_mostrar = valores[:max_mostrar] + [sum(valores[max_mostrar:])]
            return categorias_mostrar, valores_mostrar
    
    # Configurar el estilo
    plt.style.use('default')
    fig, axes = plt.subplots(4, 2, figsize=figsize)  # Cambio a 4 filas, 2 columnas
    fig.suptitle(f'{titulo_principal}\nAnálisis de {columna_categoria.upper()} por {columna_numerica.upper()}', 
                 fontsize=16, fontweight='bold', y=0.98)

    # 1. VENTAS TOTALES (Gráfico de barras) - Posición [0, 0]
    ax1 = axes[0, 0]
    categorias_vt = datos_analisis['ventas_totales']['categorias']
    valores_vt = datos_analisis['ventas_totales']['valores']
    
    # Preparar datos con "Otros" si es necesario
    categorias_mostrar_vt, valores_mostrar_vt = preparar_datos_con_otros(
        categorias_vt, valores_vt, max_categorias_vt)
    
    # Generar colores dinámicos
    colores_vt = generar_colores(len(categorias_mostrar_vt))
    
    bars1 = ax1.bar(range(len(categorias_mostrar_vt)), valores_mostrar_vt, color=colores_vt)
    ax1.set_title('Ventas Totales por Categoría', fontweight='bold', pad=20)
    ax1.set_xlabel(columna_categoria.capitalize())
    ax1.set_ylabel(f'{columna_numerica.capitalize()} (Millones)')
    ax1.set_xticks(range(len(categorias_mostrar_vt)))
    ax1.set_xticklabels(categorias_mostrar_vt, rotation=45, ha='right')
    
    # Agregar valores en las barras
    for bar, valor in zip(bars1, valores_mostrar_vt):
        ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(valores_mostrar_vt)*0.01,
                f'{valor:.2f}', ha='center', va='bottom', fontweight='bold')

    # 2. VENTAS PROMEDIO (Gráfico de barras horizontal) - Posición [0, 1]
    ax2 = axes[0, 1]
    categorias_vp = datos_analisis['ventas_promedio']['categorias']
    valores_vp = datos_analisis['ventas_promedio']['valores']
    
    # Preparar datos con "Otros" si es necesario
    categorias_mostrar_vp, valores_mostrar_vp = preparar_datos_con_otros(
        categorias_vp, valores_vp, max_categorias_vp)
    
    # Generar colores dinámicos
    colores_vp = generar_colores(len(categorias_mostrar_vp))
    
    bars2 = ax2.barh(range(len(categorias_mostrar_vp)), valores_mostrar_vp, color=colores_vp)
    ax2.set_title('Ventas Promedio por Categoría', fontweight='bold', pad=20)
    ax2.set_xlabel(f'{columna_numerica.capitalize()} Promedio (Millones)')
    ax2.set_ylabel(columna_categoria.capitalize())
    ax2.set_yticks(range(len(categorias_mostrar_vp)))
    ax2.set_yticklabels(categorias_mostrar_vp)
    
    # Agregar valores en las barras
    for bar, valor in zip(bars2, valores_mostrar_vp):
        ax2.text(bar.get_width() + max(valores_mostrar_vp)*0.01, bar.get_y() + bar.get_height()/2,
                 f'{valor:.2f}', ha='left', va='center', fontweight='bold')

    # 3. CANTIDAD DE JUEGOS (Gráfico de dona) - Posición [1, 0]
    ax3 = axes[1, 0]
    categorias_cj = datos_analisis['cantidad_juegos']['categorias']
    valores_cj = datos_analisis['cantidad_juegos']['valores']
    
    # Preparar datos con "Otros" si es necesario
    categorias_mostrar_cj, valores_mostrar_cj = preparar_datos_con_otros(
        categorias_cj, valores_cj, max_categorias_cj)
    
    # Generar colores dinámicos
    colores_cj = generar_colores(len(categorias_mostrar_cj))
    
    wedges, texts = ax3.pie(
        valores_mostrar_cj,
        labels=None,              # 👈 quitamos labels internos
        colors=colores_cj,
        startangle=90
    )
    
    # Crear leyenda con porcentajes
    total = sum(valores_mostrar_cj)
    labels_leyenda = [
    f"{cat} – {val} juegos ({val/total*100:.1f}%)"
    for cat, val in zip(categorias_mostrar_cj, valores_mostrar_cj)
    ]
    
    ax3.legend(
        wedges,
        labels_leyenda,
        title=columna_categoria.capitalize(),
        loc="center left",
        bbox_to_anchor=(1, 0.5)
    )
    
    # Crear efecto de dona
    centre_circle = plt.Circle((0,0), 0.70, fc='white')
    ax3.add_artist(centre_circle)
    ax3.set_title('Distribución de Cantidad de Juegos', fontweight='bold', pad=20)

    # 4. PARTICIPACIÓN DE MERCADO (Gráfico de barras con línea acumulativa) - Posición [1, 1]
    ax4 = axes[1, 1]
    categorias_pm = datos_analisis['participacion_mercado']['categorias']
    valores_pm = datos_analisis['participacion_mercado']['valores']
    
    # Preparar datos con "Otros" si es necesario
    categorias_mostrar_pm, valores_mostrar_pm = preparar_datos_con_otros(
        categorias_pm, valores_pm, max_categorias_pm)
    
    # Generar colores dinámicos
    colores_pm = generar_colores(len(categorias_mostrar_pm))
    
    # Gráfico de barras
    bars4 = ax4.bar(range(len(categorias_mostrar_pm)), valores_mostrar_pm, 
                    color=colores_pm, alpha=0.7)
    ax4.set_title('Participación de Mercado (%)', fontweight='bold', pad=20)
    ax4.set_xlabel(columna_categoria.capitalize())
    ax4.set_ylabel('Participación (%)')
    ax4.set_xticks(range(len(categorias_mostrar_pm)))
    ax4.set_xticklabels(categorias_mostrar_pm, rotation=45, ha='right')
    
    # Línea acumulativa
    ax4_twin = ax4.twinx()
    valores_acum = [sum(valores_mostrar_pm[:i+1]) for i in range(len(valores_mostrar_pm))]
    ax4_twin.plot(range(len(categorias_mostrar_pm)), valores_acum, 
                  color='red', marker='o', linewidth=2, markersize=6)
    ax4_twin.set_ylabel('% Acumulado', color='red')
    ax4_twin.tick_params(axis='y', labelcolor='red')
    
    # Agregar valores en las barras
    for bar, valor in zip(bars4, valores_mostrar_pm):
        ax4.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(valores_mostrar_pm)*0.01,
                f'{valor:.2f}%', ha='center', va='bottom', fontweight='bold', fontsize=9)

    # 5. TOP JUEGOS POR CATEGORÍA (Gráfico de barras agrupadas) - Posición [2, 0]
    ax5 = plt.subplot2grid((4, 2), (2, 0), colspan=2)  # Fila 3, ocupa ambas columnas
    # Preparar datos para el gráfico de barras agrupadas
    top_juegos = datos_analisis['top_juegos_por_categoria']
    
    # Aplicar límite de categorías si se especifica
    if max_categorias_grafico5 is not None:
        categorias_seleccionadas = list(top_juegos.keys())[:max_categorias_grafico5]
    else:
        categorias_seleccionadas = list(top_juegos.keys())[:4]  # Default: Top 4 categorías
    
    # Configurar posiciones para barras agrupadas
    x = np.arange(len(categorias_seleccionadas))
    width = 0.25
    
    # Crear barras para cada posición (1°, 2°, 3°)
    colores_top = ['#2ca02c', '#ff7f0e', '#d62728']  # Verde, naranja, rojo
    for i in range(3):  # Top 3 juegos por categoría
        valores = []
        nombres = []
        for categoria in categorias_seleccionadas:
            if i < len(top_juegos[categoria]['ventas']):
                valores.append(top_juegos[categoria]['ventas'][i])
                nombre_juego = top_juegos[categoria]['nombres'][i]
                nombres.append(nombre_juego[:15] + '...' if len(nombre_juego) > 15 else nombre_juego)
            else:
                valores.append(0)
                nombres.append('')
        
        bars = ax5.bar(x + i*width, valores, width, label=f'{i+1}° Lugar',
                      color=colores_top[i], alpha=0.8)
        
        # Agregar nombres de juegos en las barras
        for bar, nombre in zip(bars, nombres):
            if bar.get_height() > 0:
                ax5.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(valores)*0.01,
                        nombre, ha='center', va='bottom', rotation=45, fontsize=8)
    
    ax5.set_title(f'Top 3 Juegos por {columna_categoria.capitalize()}', fontweight='bold', pad=20)
    ax5.set_xlabel(columna_categoria.capitalize())
    ax5.set_ylabel(f'{columna_numerica.capitalize()} (Millones)')
    ax5.set_xticks(x + width)
    ax5.set_xticklabels(categorias_seleccionadas)
    ax5.legend()

    # 6. COMPARACIÓN DE EXTREMOS (Gráfico de barras comparativas) - Posición [2, 1]
    ax6 = plt.subplot2grid((4, 2), (3, 0), colspan=2)  # Fila 4, ocupa ambas columnas
    # Obtener categorías con mayor y menor rendimiento
    categorias_vt = datos_analisis['ventas_totales']['categorias']
    valores_vt = datos_analisis['ventas_totales']['valores']
    
    # Aplicar límite de categorías si se especifica
    if max_categorias_grafico6 is not None:
        num_top = min(max_categorias_grafico6 // 2, len(categorias_vt) // 2)
        num_bottom = min(max_categorias_grafico6 - num_top, len(categorias_vt) - num_top)
    else:
        num_top = 3  # Default: Top 3
        num_bottom = 3  # Default: Bottom 3
    
    # Top N y Bottom N
    top_cat = categorias_vt[:num_top]
    top_val = valores_vt[:num_top]
    bottom_cat = categorias_vt[-num_bottom:]
    bottom_val = valores_vt[-num_bottom:]
    
    # Combinar datos
    categorias_comparacion = top_cat + bottom_cat
    valores_comparacion = top_val + bottom_val
    
    # Generar colores dinámicos (verde para top, rojo para bottom)
    colores_comparacion = ['green'] * num_top + ['red'] * num_bottom
    
    bars6 = ax6.bar(range(len(categorias_comparacion)), valores_comparacion,
                    color=colores_comparacion, alpha=0.7)
    ax6.set_title(f'Comparación: Top {num_top} vs Bottom {num_bottom}', fontweight='bold', pad=20)
    ax6.set_xlabel(columna_categoria.capitalize())
    ax6.set_ylabel(f'{columna_numerica.capitalize()} (Millones)')
    ax6.set_xticks(range(len(categorias_comparacion)))
    ax6.set_xticklabels(categorias_comparacion, rotation=45, ha='right')
    
    # Agregar valores en las barras
    for bar, valor in zip(bars6, valores_comparacion):
        ax6.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(valores_comparacion)*0.01,
                f'{valor:.2f}', ha='center', va='bottom', fontweight='bold')
    
    # Agregar línea divisoria visual entre top y bottom
    if num_top > 0 and num_bottom > 0:
        ax6.axvline(x=num_top-0.5, color='black', linestyle='--', alpha=0.5)
        ax6.text(num_top/2, max(valores_comparacion)*0.9, 'TOP', ha='center', 
                fontweight='bold', fontsize=12, color='green')
        ax6.text(num_top + num_bottom/2, max(valores_comparacion)*0.9, 'BOTTOM', ha='center', 
                fontweight='bold', fontsize=12, color='red')
    
    # Ajustar layout y mostrar
    plt.tight_layout()
    plt.show()
#=======================================================
```
